```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

plt.axis([-1, 3, -1, 4])
plt.xlabel("x")
plt.ylabel("y")
plt.title("y=1+x/2")

x = np.linspace(-1, 3, 100)

# Additional axis
plt.plot([-1, 3], [0, 0], color="lightgray")
plt.plot([0, 0], [-1, 4], color="lightgray")

# Testing
plt.plot([0], [1], 'o', color="red")
plt.text(0, .5, "(0,1)", color="red")
plt.plot([2], [2], 'o', color="red")
plt.text(2, 1.5, "(2,2)", color="red")

# Your implementation
plt.plot(x, 1+x/2, color="black")
```
