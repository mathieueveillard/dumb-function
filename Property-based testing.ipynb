```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

plt.axis([0, 4, -1, 4])
plt.xlabel("x")
plt.ylabel("y")

x = np.linspace(0, 4, 100)

# Additional axis
plt.plot([0, 4], [0, 0], color="lightgray")
plt.plot([0, 0], [0, 4], color="lightgray")

# Testing
plt.plot([1], [1.5], 'o', color="red")
plt.text(1, 2, "(x1,f(x1))", color="red")
plt.plot([2], [2], 'o', color="red")
plt.text(2, 2.5, "(x2,f(x2))", color="red")
plt.plot([1, 2], [1.5, 2], color="red", lw=4)

# Your implementation
plt.plot(x, 1+x/2, color="black")
```
